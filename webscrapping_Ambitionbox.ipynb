```python
parent_element = soup.find('div', class_='companyListing__cardsContainer')

# Find all <h2> elements with the class "companyCardWrapper__companyName" within the parent element
company_elements = parent_element.find_all('h2', class_='companyCardWrapper__companyName')

# Extract and print the titles
for company_element in company_elements:
    company_title = company_element.get('title')

company_name = [title.text.strip() for title in company_elements]
print(company_name)
df = pd.DataFrame({'company':company_name})
df
```

```python
rating_divs = soup.find_all('div', class_='companyCardWrapper__companyRating')
print(rating_divs[0])
Ratings=[]
for rating_div in rating_divs:
    rating_class = rating_div['class'][1]  # Get the second class name
    rating_value = rating_class.split('-')[-1]

    rating_span = rating_div.find('span', class_='companyCardWrapper__companyRatingValue')
    if rating_span:
        rating_value = rating_span.text.strip()
        Ratings.append(rating_value)
Ratings
```

```python
action_wrapper = soup.find_all('a',class_="companyCardWrapper__ActionWrapper")
print(action_wrapper[0])
Action=[]


for action in action_wrapper:
    action_span = action.find('span', class_='companyCardWrapper__ActionCount')
    if action_span:
        action_count = action_span.text.strip()
        Action.append(action_count)

# Print all the reviews
print("Reviews:", Action)
```

```python
reviews = Action[::6]       # Reviews are at positions 0, 6, 12, etc.
salaries = Action[1::6]     # Salaries are at positions 1, 7, 13, etc.
interviews = Action[2::6]   # Interviews are at positions 2, 8, 14, etc.
jobs = Action[3::6]         # Jobs are at positions 3, 9, 15, etc.
benefits = Action[4::6]     # Benefits are at positions 4, 11, 18, etc.
photos = Action[5::6]

df['Reviews'] = reviews

df['Salaries'] =salaries
df['Interviews'] =interviews
df['Jobs'] =jobs
df['Benefits'] =benefits
df['photos'] =photos
```

```python
c_details=[]
c_type = soup.find_all('div', class_='companyCardWrapper__interLinkingWrapper')
print(c_type[0])

for detail in c_type:
    detail_span = detail.find('span', class_='companyCardWrapper__interLinking')
    if detail_span:
        detail_count = detail_span.text.strip()
        c_details.append(detail_count)

# Print all the reviews
print("company_details:", c_details)
```

# For All Pages (1-10)

```python
final_df = pd.DataFrame()
total_page = 10
for page_number in range(1,total_page+1):
    url =f'https://www.ambitionbox.com/list-of-companies?page={page_number}'
    headers = { 'user-agent':"Mozilla/5.0 ...",
                'accept': '"text/html,application...',
                'referer': 'https://...',}

    page = requests.get(url, headers= headers)
    soup = bs(page.text,'lxml')
    company = soup.find_all('div', class_='companyCardWrapper')

    len(company)
    name=[]
    c_rating=[]
    action=[]
    c_details=[]
    c_name=[]
    Action=[]
    c_details=[]
    c_type=[]
    c_age=[]
    employees=[]
    ownership=[]
    location=[]
    reviews=[]
    salaries=[]
    for i in company:
        c_name.append(i.find('h2').text.strip())
        c_rating.append(i.find('div', class_='companyCardWrapper__companyRating').text.strip())
    action_wrapper = soup.find_all('a',class_="companyCardWrapper__ActionWrapper")

    for action in action_wrapper:
        action_span = action.find('span', class_='companyCardWrapper__ActionCount')
        if action_span:
            action_count = action_span.text.strip()
            Action.append(action_count)



    c_details.append((i.find('div', class_='companyCardWrapper__interLinkingWrapper').text.strip()).split('|'))
    for details in company_details:
        if len(details) == 5:
            c_type.append(details[0].strip())
            employees.append(details[1].strip())
            ownership.append(details[2].strip())
            c_age.append(details[3].strip())
            location.append(details[4].strip())
        else:

            c_type.append(details[0].strip())
            employees.append(details[1].strip())
            ownership.append(np.nan)
            c_age.append(details[2].strip())
            location.append(details[3].strip())

    reviews = Action[::6]       # Reviews are at positions 0, 6, 12, etc.
    salaries = Action[1::6]     # Salaries are at positions 1, 7, 13, etc.
    interviews = Action[2::6]   # Interviews are at positions 2, 8, 14, etc.
    jobs = Action[3::6]         # Jobs are at positions 3, 9, 15, etc.
    benefits = Action[4::6]     # Benefits are at positions 4, 11, 18, etc.
    photos = Action[5::6]

    data ={"CompanyNames": c_name,
        "CompanyRatings": c_rating,
        "CompanyTypes": c_type,
        "Companysize": employees,
        "Ownership": ownership,
        "CompanyAge": c_age,
        "Location": location,
        "Reviews": reviews,
        "Salaries": salaries,
        "Interviews" : interviews,
        "Benefits" :benefits,
        "Jobs" : jobs,
       "Photos" : photos
      }
    df = pd.DataFrame(data)
    final_df = pd.concat([final_df, df], ignore_index=True)

final_df
```

```python
import re
def remove_more_info(Location):
    return re.sub(r'\s\+\d+ more', '', Location)  # \s : space, \+ for '+' sign, \d+ for digits


final_df['Location'] = final_df['Location'].map(remove_more_info)

print(final_df['Location'].unique())
```
